Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/cleaned_data.csv")  # Replace with actual CSV file

# --- Histogram: Age, BMI, Work Hours ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

relevant_columns = [
    "Age", "Situation maritale", "Nb pers à charge", "Profession du mari",
    "Niveau socio-économique", "Fumées de Tabouna", "nombre_des_taches",
 "BMI", "Domicile_monastir", "Domicile_sfax",
    "H_travail_sem", "Masque pour pesticides", "Bottes", "Gants", "Casquette/Mdhalla", "TAS", "Troubles cardio-respiratoires", "Troubles cognitifs", "Troubles neurologiques",
    "Troubles cutanés/phanères",
]

sns.histplot(data["Age"], bins=10, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Age Distribution")

sns.histplot(data["BMI"], bins=10, kde=True, ax=axes[1], color="green")
axes[1].set_title("BMI Distribution")

sns.histplot(data["H_travail_sem"], bins=10, kde=True, ax=axes[2], color="red")
axes[2].set_title("Work Hours Per Week Distribution")

plt.tight_layout()
plt.show()

# --- Heatmap of Correlations ---
plt.figure(figsize=(12, 8))
sns.heatmap(data[relevant_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# --- Boxplot: BMI vs. Exposure to Risks ---
plt.figure(figsize=(10, 6))
sns.boxplot(x=data["Tabagisme_Neffa"], y=data["BMI"], palette="Set2")
plt.title("BMI Distribution vs. Exposure to Smoke")
plt.xlabel("Exposure to Smoke")
plt.ylabel("BMI")
plt.show()

# --- Bar Charts: Health Issues vs. Risk Exposure ---
health_issues = ["Troubles cardio-respiratoires", "Troubles cognitifs", "Troubles neurologiques", "Troubles cutanés/phanères"]

for issue in health_issues:
    plt.figure(figsize=(8, 5))
    sns.barplot(x=data["Tabagisme_Neffa"], y=data[issue], ci=None, palette="coolwarm")
    plt.title(f"Prevalence of {issue} vs. Exposure to Smoke")
    plt.xlabel("Exposure to Smoke")
    plt.ylabel("Frequency")
    plt.show()
```
